```python
# ðŸ‡¬ðŸ‡§ Libraries
# ðŸ‡§ðŸ‡· Bibliotecas
import datetime
import altair as alt
import operator
import pandas as pd

s = pd.read_csv('../../data/prefeitura-de-aparecida/aparecida-small-without-duplicates.csv', parse_dates=['date'])
pd.options.mode.chained_assignment = None  # default='warn'

city = s[s['city'] == 'Aparecida']

base = alt.Chart(city).mark_line().encode(x = 'date').properties(width = 500)

confirmed = alt.value("#106466")
death = alt.value("#D8B08C")
recovered = alt.value("#87C232")

base.encode(y = "totalCases", color = confirmed).properties(title = "Total confirmed") | base.encode(y = "totalDeaths", color = death).properties(title = "Total deaths")
```

```python
city['new cases'] = city['totalCases'].diff()
city['new deaths'] = city['totalDeaths'].diff()
city['new recovered'] = city['totalRecovered'].diff()

# Initiate chart with data
base_cases = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new cases:Q')).properties(width = 400)
base_deaths = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new deaths:Q')).properties(width = 400)
base_recovered = alt.Chart(city).mark_bar().encode(alt.X('date:T'), alt.Y('new recovered:Q')).properties(width = 400)

# Load the chart
base_cases.encode(color = confirmed).properties(title = "Daily new cases") | base_deaths.encode(color = death).properties(title = "Daily new deaths") | base_recovered.encode(color = recovered).properties(title = "Daily new recovered")
```
